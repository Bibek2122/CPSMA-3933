```python
import random

# Function to simulate pulling a single bottle
# 10 oz 90% of the time, 9 oz 10% of the time
def pull_bottle():
    return 9 if random.random() < 0.1 else 10

# Q5 & Q6: Function to pull bottles sequentially until 9 oz is found
def sequence_until_short():
    """
    Returns a list of bottle weights pulled until the short bottle (9 oz) is found.
    """
    sequence = []  # stores the sequence of pulls
    while True:
        weight = pull_bottle()
        sequence.append(weight)
        if weight == 9:  # stop when short bottle is found
            break
    return sequence

# Example: run the function once
seq = sequence_until_short()
print("Sequence of pulls until 9 oz is found:", seq)
print("Number of pulls in this trial:", len(seq))

# Run multiple simulations to see variability and compute average pulls
num_trials = 100
trials = [sequence_until_short() for _ in range(num_trials)]
average_pulls = sum(len(s) for s in trials) / num_trials
print("Average number of pulls over 100 trials:", average_pulls)
```

```python
# Qn 5 and 6
import random

# Function to simulate pulling a single bottle
# 10 oz 90% of the time, 9 oz 10% of the time
def pull_bottle():
    return 9 if random.random() < 0.1 else 10

# Q5 & Q6: Function to pull bottles until 9 oz is found
def sequence_until_short():
    """
    Pulls bottles one at a time until a 9-oz bottle is found.
    Returns the list of weights pulled (sequence) and stops immediately when 9 is found.
    """
    sequence = []
    while True:
        weight = pull_bottle()
        sequence.append(weight)
        if weight == 9:
            break
    return sequence

# Example: run the function once
seq = sequence_until_short()
print("Sequence of pulls until 9 oz is found:", seq)
print("Number of pulls in this trial:", len(seq))

# Run multiple simulations to get average number of pulls
num_trials = 100
trials = [sequence_until_short() for _ in range(num_trials)]
average_pulls = sum(len(s) for s in trials) / num_trials
print("Average number of pulls over 100 trials:", average_pulls)
```

```python
Q5 & Q6: Function to Pull Bottles Until the Short Bottle is Found (Sequence Modification)
We create a function that pulls bottles sequentially, records each pull in a list, and stops immediately when a 9-oz bottle is found.
This allows us to see both the number of pulls and the actual sequence of pulls.
Running this function multiple times shows the random variability: some trials may find the 9-oz bottle in 1–2 pulls, others may take more.
Averaging across many trials approximates the theoretical expected number of pulls (10).

Example Output:
Sequence of pulls until 9 oz is found: [10, 10, 9]
Number of pulls in this trial: 3
Average number of pulls over 100 trials: approximately 10

Observations:
The recorded sequence highlights the randomness of each trial.
The average number of pulls over multiple trials closely matches the theoretical expectation of 10.
```

```python
Daughter rejection process.
Ensures at least one pink plate is always pulled. Non-pink rejected plates go back on top of the stack.
```

```python
def appease_daughter(stack, n=4):
    while True:
        pulled = [stack.pop() for _ in range(n)]
        if 'pink' in pulled:
            rejected = [p for p in pulled if p != 'pink']
            stack = rejected + stack  # return rejected on top
            return pulled, stack

stack = plates.copy()
pulled, stack = appease_daughter(stack)
print("Pulled plates appeasing daughter:", pulled)
```

```python
Q8: 1000-day simulation
Theory:
Tracks how often each plate color is used. Pink plates end up on top due to repeated selection.
Over 1000 days, pink plates are used most frequently because the daughter always requires at least one pink plate.
Non-pink plates are used less and often returned to the stack when no pink is present. 
As a result, pink plates migrate toward the top of the stack, while some non-pink plates remain at the bottom. 
Usage counts show which plates are regularly used and which are not.
This simulates how repeated selection affects both plate usage and stack order.
```

```python
usage_count = {'pink':0, 'blue':0, 'black':0, 'green':0}
stack = plates.copy()

for _ in range(1000):
    pulled, stack = appease_daughter(stack)
    for p in pulled:
        usage_count[p] += 1

print("Usage count after 1000 days:", usage_count)
```
